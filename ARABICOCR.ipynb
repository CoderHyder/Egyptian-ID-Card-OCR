```python
from ArabicOcr import arabicocr
import cv2
import numpy as np
import matplotlib.pyplot as plt
import json
from difflib import SequenceMatcher
from PIL import Image

def preprocess_image(image_path, scale_factor=3):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found at {image_path}")

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    contrast = cv2.convertScaleAbs(gray, alpha=2.0, beta=0.0)   # alpha = 2.0, beta = 0.0

    _, binary = cv2.threshold(contrast, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Apply dilation = adds a border around the text to make it bold
    kernel = np.ones((1, 1), np.uint8)
    dilated = cv2.dilate(binary, kernel, iterations=1)
    return dilated

#def crop_and_ocr(image, boxes, temp_image_path, annotations_path, padding = 10): # manual
def crop_and_ocr(image, boxes, temp_image_path, annotations_path):  #dynamic
    texts = []
    h, w = image.shape[:2]

    with open(annotations_path, 'w') as f:
        for i, box in enumerate(boxes):
            x_min, y_min = np.min(box[:, 0]), np.min(box[:, 1])
            x_max, y_max = np.max(box[:, 0]), np.max(box[:, 1])
            # Calculate the width and height of the bounding box
            box_width = x_max - x_min
            box_height = y_max - y_min
            #print(f"Box {i}: Width={box_width}, Height={box_height}")
            # Implement dynamic padding based on bounding box size
            # Adjust the coefficients to fine-tune the padding
            padding_x = int(0.15 * box_width)  # % of the box width (dynamic)
            padding_y = int(0.15 * box_height) # % of the box height (dynamic)
            # Ensure bounding box is within image boundaries
            x_min, y_min = max(0, x_min - padding_x), max(0, y_min - padding_y) #manual: change padding_x,y -> padding
            x_max, y_max = min(w, x_max + padding_x), min(h, y_max + padding_y)

            # Check if the box is valid
            if x_min < x_max and y_min < y_max:
                # Crop the image based on the bounding box
                cropped_img = image[y_min:y_max, x_min:x_max]

                # Save the cropped image
                crop_filename = f'{temp_image_path}{i}.png'
                cv2.imwrite(crop_filename, cropped_img)

                # Perform OCR on the cropped image
                result = arabicocr.arabic_ocr(crop_filename, crop_filename)
                text = result[0][1].strip() if result else ""
                texts.append(text)

                # Write the result to the annotations file
                f.write(f'{text}, {crop_filename}, {box.tolist()}\n')
            else:
                texts.append("")
                f.write(f'Invalid bounding box, {box.tolist()}\n')

    return texts

def extract_text_with_boxes(image_path, output_path):
    # Preprocess the image
    preprocessed_image = preprocess_image(image_path)
    preprocessed_image_path = 'preprocessed_image.png'
    cv2.imwrite(preprocessed_image_path, preprocessed_image)

    # Run OCR to get the bounding boxes and associated text
    result = arabicocr.arabic_ocr(preprocessed_image_path, output_path)
    print("Result:", result)

    # Sort by y-coordinate (top to bottom) first, then by x-coordinate (right to left)
    sorted_result = sorted(result, key=lambda item: (item[0][0][1], -item[0][0][0])) # 001 000

    # Extract bounding boxes from the sorted result
    boxes = [np.array(item[0]).astype(int) for item in sorted_result]

    # Define a temporary image path for saving cropped images
    temp_image_path = "/content/temp_cropped"
    annotations_path = "/content/annotations.txt"

    # Crop and perform OCR on each cropped image
    texts = crop_and_ocr(preprocessed_image, boxes, temp_image_path, annotations_path)

    print("Processing file:", image_path)
    print("Extracted Texts:", texts)

    return boxes, texts

def display_with_boxes(image_path, boxes, texts, scores):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    for box, text, score in zip(boxes, texts, scores):
        box = np.array(box).astype(int)
        cv2.polylines(image, [box], isClosed=True, color=(255, 255, 255), thickness=2)
        cv2.putText(image, f'{text} ({score:.2f})', (box[0][0], box[0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def process_image(image_path, output_path):
    boxes, texts = extract_text_with_boxes(image_path, output_path)
    display_with_boxes(image_path, boxes, texts, [1.0] * len(texts))  # Dummy scores for display

def process_and_resize_image(input_path, crop_coords, output_dir, output_size):
    # Define output file paths
    cropped_filename = f'{output_dir}/cropped_nid_back_image.jpg'
    high_res_filename = f'{output_dir}/high_res_nid_back_image.jpg'

    # Open the original image
    img = Image.open(input_path)

    # Crop the image
    img_crp = img.crop(crop_coords)
    if img_crp.mode == 'RGBA':
        img_crp = img_crp.convert('RGB')
    img_crp.save(cropped_filename)
    img_crp.show()

    # Print cropped image size
    cropped_size = img_crp.size
    print(f'Cropped image size: {cropped_size}')

    # Resize the cropped image
    img_high_res = img_crp.resize(output_size)
    img_high_res.save(high_res_filename)
    img_high_res.show()

    # Print high-resolution image size
    high_res_size = img_high_res.size
    print(f'High-resolution image size: {high_res_size}')

    return high_res_filename

input_path = '/content/nid_back_4.jpg'
crop_coords = (236, 6, 925, 395)  #(236, 6, 925, 395) back. front -> (67,210,1094,695)
output_dir = '/content'
output_size = (1200, 768)       #(1350, 650) front back #(1200, 768)

high_res_path = process_and_resize_image(input_path, crop_coords, output_dir, output_size)
print("High res path name:", high_res_path)
output_path = '/content/output.png'

process_image(high_res_path, output_path)
```
